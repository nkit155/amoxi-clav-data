load the dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Amoxi-clav_Sales_Data_Sample.csv")
df
```

```python
df.info()
```

convert format of date to object to datetime

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.info()
```

create new column for year and month from date column

```python
df['year']=df['Date'].dt.year
df['month']=df['Date'].dt.month_name()
df
```

creat new column for sales and convert format of sales column from float to integar

```python
df['sales_value']=df['Sales Value (INR)']
df['sales_value']=df['sales_value'].astype(int)
df
```

delete th column of sales value(INR)

```python
df=df.drop(columns=['Sales Value (INR)'])
df
```

find out total company sales by apply the aggregate function

```python
company_sales=df.groupby('Company').agg(total_sales=('sales_value','sum'))
company_sales
```

find out top 5 company sales by sort values(desending) and also apply the head function

```python
top_5_company=company_sales.sort_values('total_sales',ascending=False).head(5).reset_index()
top_5_company
```

top product SKU range wise sales

```python
prod_sku=df.groupby('Product Name').agg(sku_sales=('sales_value','sum')).sort_values('sku_sales',ascending=False).reset_index()
prod_sku
```

sales of every products in every company 

```python
comp_prod_sales=df.groupby(['Company','Product Name']).agg(prod_sales=('sales_value','sum')

)
comp_prod_sales
```

company wise product sales in descending order

```python
comp_prod_sales = comp_prod_sales.sort_values(['Company', 'prod_sales'], ascending=[True, False])
comp_prod_sales
```

```python
comp_prod_sales.to_excel('comp_prod_sales.xlsx')
```

 top month wise product sale

```python
month_sales=df.groupby('month').agg(mon_sales=('sales_value','sum')

).sort_values('mon_sales',ascending=False)
month_sales
```

```python
month_prod_sales=df.groupby(['month','Product Name']).agg(mon_sales=('sales_value','sum')
                                    ).sort_values(['month','mon_sales'],ascending=[True,False])
month_prod_sales
```

company based  unit_sales

```python
company_unt=df.groupby('Company').agg(total_unit=('Sales Units','sum')

).sort_values('total_unit',ascending=False)
company_unt
```

top 5 company based unit sales

```python
top_5_comp_unit=company_unt.head(5).reset_index()
top_5_comp_unit
```

month base total unit

```python
month_unit_sales=df.groupby('month').agg(month_unit=('Sales Units','sum')

).sort_values('month_unit',ascending=False)
month_unit_sales
```

unit of every product in every month

```python
mon_prod_unit_sales=df.groupby(['month','Product Name']).agg(total_mon_prod=('Sales Units','sum')

)
mon_prod_unit_sales
```

```python
sort_mon_prod_unit_sales=mon_prod_unit_sales.sort_values(['month','total_mon_prod'],ascending=[True,False])
sort_mon_prod_unit_sales
```

average unit of every product of every month

```python
avg_mon_prod_unit_sales=df.groupby(['month','Product Name']).agg(mon_avg_prod=('Sales Units','mean')

)
avg_mon_prod_unit_sales
```

sales of every company of every month

```python
comp_mon_sales=df.groupby(['month','Company']).agg(total_comp_sales=('sales_value','sum')

).reset_index()
comp_mon_sales
```
